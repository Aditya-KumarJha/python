# Logging with multiple Loggers
### We can create multiple loggers for different parts of our application.

```python
import logging
```

```python
## create a logger for module 1
logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## create a logger for module 1
logger2 = logging.getLogger("module2")
logger2.setLevel(logging.WARNING)
```

```python
## Configure logging settings
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
## log messages with different loggers
logger1.debug("This is a debug message for module1")
logger2.warning("This is a warning message for module2")
logger2.error("This is an error message for module2")
```
