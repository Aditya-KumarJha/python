# Importing Modules in Python: Modules and Packages

### In Python, modules and packages help organize code and reuse code. Here's a comprehensive guide on how to import them.

```python
import math
math.sqrt(16)
```

```python
from  math import sqrt, pi
print(sqrt(35))
print(pi)
```

```python
import numpy as np ## np is alias for numpy
np.array([1, 2, 3, 4])
```

```python
from math import *
print(sqrt(25))
print(pi)
```

```python
from package.maths import addition
addition(8, 13)
```

```python
from package import maths
print(maths.addition(8, 19))
print(maths.addition(8, 29))
print(maths.subtraction(8, 21)) 
```
