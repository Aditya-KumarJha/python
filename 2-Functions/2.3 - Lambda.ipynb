# Lambda Functions in Python

### Lambda functions are small anonymous functions defined using the lambda keyword. They can have any nummber of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions.

```python
# Syntax
# lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(4,9)
```

```python
add = lambda a,b: a+b
type(add)
print(add(8,13))
```

```python
def even(num):
    if not (num%2):
        return True
    
even(28)
```

```python
even1 = lambda num: num%2 == 0
even1(17)
even1(16)
```

```python
def addition(x, y, z):
    return x + y + z

addition(88, 123, 96)
```

```python
addition1 = lambda x, y, z : x + y + z
addition1(88, 987, 465)
```

```python
## map() ->  It applies a function to all items in a list. It requires first parameter as function and second as iterable

numbers = [1, 2, 3, 4, 5, 6]

def square(number):
    return number * number

square(5)
```

```python
list(map(lambda x : x**2, numbers))  ## map uses lasy loading technique
```
