```python
def even_or_odd(num):
    """This functions find even or odd"""
    if num%2:
        print("The number is odd")
    else:
        print("The number is even")
```

```python
## call function
even_or_odd(11)
```

```python
## Function with multiple parameters
def add(a,b):
    return a+b

result = add(4,7)
print(result)
```

```python
## Default parameters

def greet(name="Guest"):
    print(f"Hello {name}")

greet("Aditya")
greet()
```

```python
### Variable Length Arguments
## Positional and Keywords arguments

## Positional Arguments
def print_numbers(*args) : ## *args refers to all positional arguments
    for num in args:
        print(num)

print_numbers(1, 2, 3, 4, 5, 6, 7, 8, "Aditya")
```

```python
## Keywords Arguments

## Here all the argument will be in the form of key - value pair
def print_details(**kwargs):
    for key, val in kwargs.items():
        print(key, val)

print_details(name="Aditya", age=32, country="India")
```

```python
## combine both
def print_details(*args, **kwargs):
    for num in args:
        print("Positional argument :",num)

    for key, val in kwargs.items():
        print(key, val)

print_details(1,2,3,"Krish", name="Aditya", age=32, country="India") # First all positional arguments they keyword arguments
```

```python
### return statements
## with return statement our function return some value(s)

def mul(a,b):
    return a*b

print(mul(5,8))
```
