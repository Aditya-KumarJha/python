# Iterators

### Iterators are advanced Python concepts that allows for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
myList = [1, 2, 3, 4, 5, 6]
for i in myList:
    print(i)

type(myList)
```

```python
## Iterator

iterator = iter(myList)
print(type(iterator))
```

```python
## it uses lazy loading technique it means that list_iterator is created at some memory location and it will not give us any element until and unless we iterate through that particular element
iterator 
```

```python
## iterate through all element
next(iterator)
```

```python
iterator = iter(myList)
```

```python
try:
   print(next(iterator))
except StopIteration:
    print("There are no more elements in iterator")
```

```python
# String iterator
myString = "Hello"
stringIterator = iter(myString)

print(next(stringIterator))
print(next(stringIterator))
```
