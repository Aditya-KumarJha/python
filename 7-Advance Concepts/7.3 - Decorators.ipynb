# Decorators

### Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. 

```python
### function copy
### closures
### decorators
```

```python
## function copy

def welcome() :
    return "Welcome to advanced python concepts"

welcome()
```

```python
## function copy

wel = welcome  ##  we can copy function and even then if we delete we will have it's reference copy

print(wel)
print(wel())

del welcome
print(wel())
```

```python
## closures => function inside a function

def mainWelcome(check):
    msg = "Welcome"
    def subWelcome():
        print("Welcome to advance python concepts")
        print(msg)  ## we can also access anything which is defined outside sub-method
        print(check)
        print("Learn all concepts properly")
    
    return subWelcome()
```

```python
mainWelcome("Hii")
```

```python
def mainWelcome(func):

    def subWelcome():
        print("Welcome to advance python concepts")
        func("Welcome everyone")
        print(func)
        print("Learn all concepts properly")
    
    return subWelcome()
```

```python
mainWelcome(print)
```

```python
def mainWelcome(func, lst):

    def subWelcome():
        print("Welcome to advance python concepts")
        print(func)
        print(func(lst))
        print("Learn all concepts properly")
    
    return subWelcome()
```

```python
mainWelcome(len, [1, 2, 3, 4, 5, 6])
```

```python
## Decorator

def mainWelcome(func):

    def subWelcome():
        print("Welcome to advance python concepts")
        func()
        print(func)
        print("Learn all concepts properly")
    
    return subWelcome()
```

```python
def courseIntro():
    print("Advance and important concepts to master")

courseIntro()
```

```python
mainWelcome(courseIntro)
```

```python
@mainWelcome ## @ for decorator so when we wrote mainWelcome as decorator means the other function was going under that as a parameter
def courseIntro():
    print("Advance and important concepts to master")
```

```python
def myDecorator(func):
    def wrapper():
        print("Something is happening before the function is called")
        func()
        print("Something is happening after the function is called")

    return wrapper
```

```python
@myDecorator
def sayHello():
    print("Hello!")
    
```

```python
sayHello()
```

```python
## Decorators with arguments

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n): # The _ is just a conventional throwaway variable in Python. It’s used when: We need to repeat something n times, but we don’t care about the loop index
                func(*args, **kwargs)

        return wrapper
    return decorator
```

```python
@repeat(3)
def sayHello():
    print("Hello")
```

```python
sayHello()
```

## Conclusion

### Decorators are a powerful tool in Python for extending and modifying the behavior of functions and methods. They provide a clean and readable way to add functionality such as logging, timing, access control and more without changing the original code.
