# Python Memory Management

### Memory Management in Python involves a combination of automatic garbage collection, reference counting, and various internal optimiations to efficiently manage memory allocation and deallocation. Understanding these mechanisms can help developers write more efficient and robust applications.

1. Key concepts in Python Memory Management:
   - Memory Allocation and Deallocation
   - Reference Counting
   - Garbage Collection
   - The gc module
   - Memory management Best Practices
   

## Reference COunting

### Reference counting is the primary method Python uses to manage memory. Each object in Python maintains a count of references pointing to it. When the reference count drops to zero, the memory occupied by the object is deallocated

```python
import sys

a = []
## 2 (one reference from 'a', one from getrefcount's argument)
print(sys.getrefcount(a))
```

```python
b = a
print(sys.getrefcount(b))
```

```python
del b
print(sys.getrefcount(a))
```

## Garbage Collection

### Python includes a garbage collector to handle reference cycles. Reference cycles occur when objects reference each other, preventing their reference count from reaching zero.

```python
import gc

## enable garbage collection
gc.enable()
```

```python
## disable garbage collection
gc.disable()
```

```python
## manually trigger garbage collection
gc.collect()
```

```python
### Get garbage collector statistics
print(gc.get_stats())
```

```python
### get unreachable objects
print(gc.garbage)
```

## Memory Management Best Practices

1. Use Local Variables: Local variables have a shorter lifespan and are freed sooner than global variables.
2. Avoid Circular References: Circular references can lead to memory leaks if not properly managed.
3. Use Generators: Generators produce items one at a time and only keep one item in memory at a time, making the mory efficient.
4. Explicitly Delete Objects: Use the del statement to delete variables and objects explicitly when they are no longer needed.
5. Profile Memory Usage: Use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.

```python
### Handling Circular References

import gc

class MyObject:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created")
    
    def __del__(self):
        print(f"Object {self.name} destroyed")

## creating circular reference
obj1 = MyObject("obj1");
obj2 = MyObject("obj2");
obj1.ref = obj2
obj2.ref = obj1

## delete objects
del obj1
del obj2

## force garbage collection / manually trigger garbage collection
gc.collect() # If not present then objects will not be destroyed
```

```python
## Generators For Memory Efficiency
### Generators allow us to produce items one at a time, using memory efficiently by only keeping one item in memory at a time.

def generate_numbers(n):
    for i in range(n):
        yield i

## using the generator
for num in generate_numbers(1000000):
    print(num)
    if num > 10:
        break
```

```python
## Profiling Memory Usage with tracemalloc

import tracemalloc

def create_list():
    return [i for i in range(1000000)]

def main():
    tracemalloc.start()

    create_list()

    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')

    print("Top 10 memory allocations:")
    for stat in top_stats[:10]:
        print(stat)

main()
```
