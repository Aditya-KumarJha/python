# FLASK FRAMEWORK

## WSAI -> Web Server Gateway Interface
## JINJA2 -> Template Engine

## FLASK : Complete Web Framework which is created with the Python Programming Language

## WSGI : Web Server Gateway Interface used to build web applications with Python. It helps in communication between web servers and web applications.
## JINJA2 : Template Engine used by Flask to render HTML templates. Basically Layout of a page is combined with data source to generate dynamic web pages.
