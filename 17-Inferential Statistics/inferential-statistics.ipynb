# Inferential Statistics: It refers to the process of making inferences or drawing conclusions about a population based on information obtained from a sample.

## Estimate: Estimate is specified observed numerical value used to estimate an unknown population parameter. Different types of estimates are:
- **Point Estimate**: A single numerical value used to estimate an unknown population parameter. E.g., sample mean is a point estimate of population mean.
- **Interval Estimate**: A range of values used to estimate an unknown population parameter. E.g., sample mean ± margin of error.

## Hypothesis Testing Mechanism

### NULL Hypothesis (H₀): The assumptions we are beginning with
### ALTERNATIVE Hypothesis (H₁): The claim we are testing i.e., opposite of H₀
### Experiment: Collect data to test the hypothesis (Statistical Analysis)
### Accept or Reject H₀ based on the evidence


### P Value: The p value is a number, calculated from a statistical test, that describes how likely you are to have found a particular set of observations if the null hypothesis were true. P values are used in hypothesis testing to help decide whether to reject the null hypothesis.

### Statistical Analysis
- Z-Test (Used when population standard deviation is known and n >= 30) / Deals with Average data
- T-Test (Used when population standard deviation is unknown and n < 30) / Deals with Average data
- ANOVA also called (Analysis of Variance) (Used to compare means of more than two groups) / Deals with Average data 
- Chi-Square Test (Used to test association between categorical variables) / Deals with Categorical data

### Definitions
#### Z-Test
- It is a statistical test that uses the standard normal distribution to determine whether to reject the null hypothesis.
- Used when population standard deviation is known and n >= 30
- Deals with Average data
- Importaant terms:
  - **Null Hypothesis (H₀)**: The default assumption that there is no effect or no difference.
  - **Alternative Hypothesis (H₁ or Hₐ)**: The hypothesis that there is an effect or a difference.
  - **Significance Level (α)**: The probability of rejecting the null hypothesis when it is true (Type I error). Common values are 0.05 (5%) or 0.01 (1%).
  - **Test Statistic**: A value calculated from the sample data to test the null hypothesis.
  - **Critical Value**: The value that the test statistic must exceed to reject the null hypothesis.
  - **P-Value**: The probability of obtaining a test statistic as extreme as, or more extreme than, the observed value, assuming the null hypothesis is true.

#### T-Test
- It is a statistical test that uses the t-distribution to determine whether to reject the null hypothesis.
- Used when population standard deviation is unknown and n < 30
- Deals with Average data
- Importaant terms:
  - **Degrees of Freedom (df)**: The number of independent pieces of information that go into the estimate of a parameter. For a single sample t-test, df = n - 1.

#### ANOVA (F-Test)
- It is a statistical test that uses the F-distribution to determine whether to reject the null hypothesis.
- Used to compare means of more than two groups
- Deals with Average data
- Important terms:
  - **F-Statistic**: The ratio of the variance between groups to the variance within groups.
  - **Degrees of Freedom (df)**: The number of independent pieces of information that go into the estimate of a parameter. For ANOVA, df_between = k-1 and df_within = N-k where k is the number of groups and N is the total number of observations.
- Types:
  - **One-Way ANOVA**: Compares means of three or more groups based on one independent variable. It has one factor with atleast 2 levels, these levels are independent of each other.
  - **Repeated Measures ANOVA**: One factor with atleast 2 levels, these levels are dependent on each other.
  - **Factorial ANOVA**: Two or more factors with atleast 2 levels each, these levels can be independent and dependent.

#### Chi-Square Test
- It is a statistical test that uses the chi-square distribution to determine whether to reject the null hypothesis.
- Used to test association between categorical variables
- Deals with Categorical data
- Importaant terms:
  - **Degrees of Freedom (df)**: The number of independent pieces of information that go into the estimate of a parameter. For a chi-square test, df = (r-1)(c-1) where r is the number of rows and c is the number of columns.

### Formulae

#### Z-Test
$$Z = \frac{\bar{x}-\mu}{\sigma/\sqrt{n}}$$

#### T-Test
$$t = \frac{\bar{x}-\mu}{s/\sqrt{n}}$$

#### ANOVA (F-Test)
$$ F= \frac{MSB}{MSW}$$
where
$$MSB = \frac{SSB}{df_B}\qquad\text{and}\qquad MSW=\frac{SSW}{df_W}$$

#### Chi-Square Test
$$ \chi^2 = \sum \frac{(O-E)^2}{E}$$
where
$$O = \text{Observed frequency}$$
$$E = \text{Expected frequency}$$


## Type 1 and Type 2 Errors

### Reality: NULL Hypothesis is True or NULL Hypothesis is False
### Decision: NULL Hypothesis is True or NULL Hypothesis is False

#### Outcome 1: We reject the NULL Hypothesis when in reality it is false (Correct Decision)
#### Outcome 2: We reject the NULL Hypothesis when it is actually true (Type 1 Error)
#### Outcome 3: We fail to reject the NULL Hypothesis when it is actually false (Type 2 Error)
#### Outcome 4: We fail to reject the NULL Hypothesis when in reality it is true (Correct Decision)

### Confidence Interval and Margin of Error

#### Confidence Interval
A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It is calculated using the sample statistic and the margin of error. 
$$\text{Confidence Interval} = \text{Point Estimate} \pm \text{Margin of Error}$$
$$\hat{\theta} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$$

#### Margin of Error
The margin of error is the range of values above and below the sample statistic that is used to estimate the population parameter. It is calculated using the standard error of the sample statistic and the critical value from the standard normal distribution.
