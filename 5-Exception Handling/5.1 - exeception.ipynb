# Understanding Exception

### Exception Handling in Python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program.

## What are Exceptions?

### Exceptions are events that disruots the normal flow of a program. They occur when an error is encountered during program execution. Common exceptions include:
    ZeroDivisionError: Dividing by zero.
    FileNotFoundError: File not found.
    ValueError: Invalid value.
    TypeError: Invalid type.

```python
## Exception try, except block
try:
    a = b
except:
    print("Variable isn't assigned")
```

```python
try:
    a = b
except NameError as ex:
    print(ex)
```

```python
result = 1 / 0
```

```python
try:
    result = 1/0
except ZeroDivisionError as ex:
    print(ex)
    print("Please enter valid denominator")
```

```python
try:
    num = int(input("Enter a number"))
    result = 10 / num

except ValueError:
    print("This is not a valid number")

except ZeroDivisionError:
    print("Please enter valid denominator")

except Exception: 
    print('Main exception class: Can catch all exceptions')
```

```python
## try, except, else block

try:
    num = int(input("Enter a number"))
    result = 10 / num

except ValueError:
    print("This is not a valid number")

except ZeroDivisionError:
    print("Please enter valid denominator")

except Exception: 
    print('Main exception class: Can catch all exceptions')

else:
    print(f"The result is {result}")  ## if none exception is then else block will execute otherwise exception block will execute
```

```python
## try, except, else and finally block

try:
    num = int(input("Enter a number"))
    result = 10 / num

except ValueError:
    print("This is not a valid number")

except ZeroDivisionError:
    print("Please enter valid denominator")

except Exception: 
    print('Main exception class: Can catch all exceptions')

else:
    print(f"The result is {result}")  

finally:
    print("Execution complete")  ## finally block will always execute irrespective of exception found or not
```

```python
### File Handling and Exception Handling

try:
    file = open('example.txt', 'r')
    content = file.read()
    print(content)

except FileNotFoundError:
    print("The File doesn't exist")

finally:
    if 'file' in locals() and not file.closed(): ## Return a dictionary containing the current scope's local variables
        file.close()
        print('File closed')
```
