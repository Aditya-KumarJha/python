# Data Visualization With Seaborn
### Seaborn is a Python visualization library on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualization with just a few lines of code.

```python
import seaborn as sns

## Seaborn = Makes beautiful and statistically helpful plots easily

## Matplotlib = Controls the canvas, layout, and rendering
```

```python
### Basic Plotting with Seaborn

tips = sns.load_dataset('tips') # tips is a built-in sample dataset included with Seaborn.
tips
```

```python
## create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots
## Bar Plot

sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar Plot of Total Bill by day")
plt.show()
```

```python
## Box Plot

sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
## Violin Plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
### Histograms

sns.histplot(tips['total_bill'], bins=10, kde=False)
```

```python
sns.histplot(tips['total_bill'], bins=10, kde=True) # KDE (Kernel Density Estimate) curve along with the histogram.
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'])
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'], fill=True)  ## shade is now depreceated so use fill
```

```python
# Pairplot

sns.pairplot(tips)  ## use to show relationship between each and every feature. Categorical variable not taken into consideration
```

```python
tips
```

```python
# [] won't work as it's equivalent to [()] and it'll be considered tuple. And Pandas doesn't recognize a tuple of strings as valid for selecting multiple columns — hence the KeyError.
corr = tips[['total_bill', 'tip', 'size']].corr() 
corr
```

```python
### Heatmap
## annot=True displays numerical value inside square box 
## cmap=coolwarm uses coolwarm color notation generally, blue(negative), red(positive), white(neutral, around 0)
# check documentation for more colours

sns.heatmap(corr, annot=True, cmap='coolwarm') 
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## Plot total sales by product

plt.figure(figsize=(10, 6)) # unit inches width: 10 inches, height: 6 inches
sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator=sum) # in estimator we can put any aggregator function
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Plot total sales by Region

plt.figure(figsize=(10, 6)) 
sns.barplot(x='Region', y='Total Revenue', data=sales_df, estimator=sum) 
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
