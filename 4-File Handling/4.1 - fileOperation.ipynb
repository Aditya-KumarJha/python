# File Operation- Read And Write Files

### File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary.

```python
### Read a whole file

with open('example.txt', 'r') as file: ## here we created object of file
    content = file.read()
    print(content)
    
```

```python
### Read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) ## .strip() it removes newline
```

```python
## Writing a file(Overwriting)

with open('example.txt', 'w') as file:
    file.write('Hello Guysss\nMy name Aditya Kumar Jha\nA Full Stack Developer now diving into the world of AI and ML\n')
    file.write('Currently Learning ML\n')
```

```python
## Write a file (without overwrite)

with open('example.txt', 'a') as file:
    file.write("Hope I do Well\n")
```

```python
### Writing a list of lines to a file

lines = ['First line \n', 'Second line \n', 'Third line \n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary Files

## Writing to a binary file

data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file: ## write byte mode
    file.write(data)
```

```python
## Reading a binary file

with open('example.bin', 'rb') as file: 
    content = file.read()
    print(content)
```

```python
### Read the content from source text file and write to destination text file

with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
## Read a text file and count no of lines, words, and characters

def countTextFile(filePath):
    with open(filePath, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = "example.txt"
lines, words, characters = countTextFile(file_path)
print("Line Count ",lines, " Word Count ",words, " Character Count ", characters)
```

### The w+ mode in Python is used to open a file for both reading and writing. If the file doesn't exist it will be created. If file exists it's content is truncated(i.e., the file is overwritten)

```python
### writing and then reading a file

with open ('example.txt', 'w+') as file:
    file.write('Hello Guysss\nMy name Aditya Kumar Jha\nA Full Stack Developer now diving into the world of AI and ML\n')
    file.write('Currently Learning ML\n')

    ## Move the file cursor to the beginning to read
    file.seek(0)

    ## Read the content of the file
    content = file.read()
    print(content)
```
