# Encapsulation and Abstraction

### Encapsulation and abstraction are two fundamental principles of Object-Oriented Programming (OOP) that helps in designing robust, maintainable and reusable code. Encapsulation involves bundling data and methods that operates on the data within a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features.

## Encapsulation

### Encapsulation is the concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access to some of the object's components, which is a mean of preventing accidental interference and misuse of the data.

```python
### Encapsulation we can do with getter and setter method
## public, protected, private variables or access modifiers

class Person:
    def __init__(self, name, age):
        self.name = name  ## these instance variables are public variables
        self.age = age

def getName(person):
    return person.name

person = Person("Aditya", 21)
getName(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name  ## by using these __ variables become private. We can't access even in derived class
        self.__age = age
        self.gender = gender ## this is public
```

```python
person = Person("Aditya", 21, "Male")
dir(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self._name = name  ## by using these __ variables become protected
        self._age = age
        self.gender = gender ## this is public

# can't access protected from outside class but can be accessed from derived class

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)
```

```python
employee = Employee("Aditya", 21, "Male")
print(employee._name)
```

```python
## Encapsulation with Getter and Setter
class Person:
    def __init__(self, name, age):
        self.__name = name  
        self.__age = age

    ## getter method for name
    def getName(self):
        return self.__name
    
    ## getter method for age
    def getAge(self):
        return self.__age
    
    ## setter method for name
    def setName(self, name):
        self.__name = name
    
    ## setter method for age
    def setAge(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")
    
```

```python
person = Person("Aditya", 21)

## access and modify private variable using getter and setter
print(person.getName())
print(person.getAge())

person.setAge(20)
person.setName("Adi")

print(person.getName())
print(person.getAge())

person.setAge(-2)
```
