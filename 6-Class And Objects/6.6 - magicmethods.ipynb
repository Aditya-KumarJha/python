# Magic Methods

### Magic methods in Python, also known as dunder methods (double underscore methods), are special methods that start and end with double underscores. These methods enable you to define the behavior of objects for built-in operations, such as arithmetic operations, comparisons and more.

### Magic methods are predefined methods in Python that we cab override to change the behavior of our objects. Some common magic methods include:

```python
'''

__init__: Initializes a new instance of a class.
__str__: Return a string representation of an object.
__repr__: Return an official string representation of an object.
__len__: Returns the length of an object.
__getitem__: Gets an item from a container.
__setitem__: Sets an item in a container.

'''
```

```python
## Basics Methods

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("Aditya", 21)
print(person)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, is {self.age} years old"
    
    def __repr__(self):
        return f"Person(name = {self.name}, age = {self.age})"

person = Person("Aditya", 21)
print(person)
print(repr(person))
```
