# Custom Exception (Raise and Throw an Exception)

```python
class Error(Exception) :
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Enter the dob"))
age = 2025 - year

try:
    if age <= 30 and age >=20:
        print("Valid age to apply for exams")

    else:
        raise dobException
    
except dobException:
    print("Sorry your age doesn't fit in the guideline")
```
