# Inheritance In Python

### Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes and methods from another class.

```python
## Inheritance

# Parent Class

class Car:
    def __init__(self, windows, doors, engineType):
        self.windows = windows; self.doors = doors
        self.engineType = engineType

    def drive(self):
        print(f"The person will drive the {self.engineType} car")
```

```python
car1 = Car(4, 5, "Petrol")
car1.drive()
```

```python
## Single Inheritance Child class inheriting from ne parent class

class Tesla(Car):   # Whatever class name we write inside () means we are inheriting from that class
    def __init__(self, windows, doors, engineType, isSelfDriving):
        super().__init__(windows, doors, engineType) ## it means call parent class constructor
        self.isSelfDriving = isSelfDriving ## this attribute is only in child class

    def selfDriving(self):
        print(f"Tesla supports self driving : {self.isSelfDriving}")
```

```python
tesla1 = Tesla(4, 5, "Electric", True)
tesla1.drive()
tesla1.selfDriving()
```

```python
### Multiple Inheritance
## When a class inherits from more than one base class.

## Base Class 1
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Subclass must implement this method")

## Base Class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner

    def speak(self):
        print("Subclass must have this method")
```

```python
# Derived class
class Dog(Animal, Pet):  ## Python will call the first speak method it finds in the MRO (left-to-right order of base classes)
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
    
    def speak(self):
        return f"{self.name} says woof"
    
# Create an object
dog = Dog("Buddy", "Aditya")
dog.speak()
```
