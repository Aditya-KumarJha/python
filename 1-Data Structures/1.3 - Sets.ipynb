### Sets

    Sets are built-in data type in python used to store collection of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership test, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
## create a set
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1, 2, 3, 4, 5, 6])
print(my_set)
```

```python
my_set = set([1, 2, 3, 6, 5, 4, 4, 5, 6])
print(my_set)
```

```python
## Basics Sets Operation
## Adding and removing elements

# Adding element
my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)

# Removing element
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10) ## error as element isn't present
```

```python
my_set.discard(10)  ## won't throw error like remove if element is not present and if it's present it'll simply discard that
print(my_set)
```

```python
## pop method

removed_element = my_set.pop()
print(removed_element)
```

```python
## clear all the elements

print(my_set)
my_set.clear()
print(my_set)
```

```python
## Set Membership Test
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)  ## True as element is present
print(10 in my_set) ## False as element is absent
```

```python
## Mathematical Operation

set_1 = {1, 2, 3, 4, 5, 6}
set_2 = {4, 5, 6, 7, 8, 9}

### Union
union_set = set_1.union(set_2)
print(union_set)

### Intersection
intersection_set = set_1.intersection(set_2)
print(intersection_set)

set_1.intersection_update(set_2)    ## Found common element and updated set_1 value with those common elements
print(set_1)
```

```python
set_1 = {1, 2, 3, 4, 5, 6}
set_2 = {4, 5, 6, 7, 8, 9}

### Difference
print(set_1.difference(set_2))
print(set_2.difference(set_1))
```

```python
### Symmetric Difference

print(set_1.symmetric_difference(set_2)) ## Common elements are removed from both set and we get new set of all unique elements from both sets
print(set_2.symmetric_difference(set_1))
```

```python
## Sets Methods

set_1 = {1, 2, 3}
set_2 = {3, 4, 5}

## is subset
print(set_1.issubset(set_2))
print(set_1.issuperset(set_2))
```

```python
## Sets Methods

set_1 = {1, 2, 3, 4, 5}
set_2 = {3, 4, 5}

## is subset
print(set_1.issubset(set_2))
print(set_1.issuperset(set_2))
```

```python
lst = [1, 2, 2, 3, 4, 4, 5]
set(lst)
```

```python
text = "In this tutorial we are discussing about sets"
words = text.split() ## split by spaces

## convert list of words to set to get unique words
unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
