### Dictionaries
    Dictionaries are unordered collection of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., string, numbers, tuples), while value can be of any type. Dictionary are mutable i.e., we can add, update or delete elements.

```python
## Creating Dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name": "Aditya", "age":21, "grade": 24}
print(student)
print(type(student))
```

```python
# Single Key is always used and takes recent value
student = {"name": "Aditya", "age":21, "name": 24}
print(student)
```

```python
## Accessing dictionary elements
student = {"name": "Aditya", "age":21, "grade": 'A'}
print(student)
print(student['grade'])
print(student['age']) 

## Accessing using get() method
print(student.get('grade'))
print(student.get('last'))
print(student.get('last', "Not Available"))  ## Default value if we don't find any element corresponding to mentioned key
```

```python
## Modifying Dictionary Elements
print(student)
```

```python
student["age"] = 33  ## updated value for key
print(student)
student["address"] = "India" ## added new value and key
print(student)
```

```python
del student['grade']  ## delete key and value pair
print(student)
```

```python
## Dictionary methods

keys = student.keys()   ## will get all the keys
print(keys)
values = student.values() ## will get all the values
print(values)

items = student.items() ## get all key value pairs in list of tuples
print(items)
```

```python
## shallow copy

student_copy =  student
print(student)
print(student_copy)
```

```python
student["name"] = "Adi"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## shallow copy it will allocate different memory i.e, it'll take variable and will give reference to another memory
print(student_copy1)
print(student)
```

```python
student["name"] = "Aditya"
print(student_copy1)
print(student)
```

```python
### Iterating over dictionaries
## We can use loops to iterate over dictionaries keys or values or items

## Iterating over keys
for keys in  student.keys():
    print(keys)

print()

## Iterating over values
for val in  student.values():
    print(val)

print()

## Iterating over items
for key,val in  student.items():
    print(f"Key -> {key} : Value -> {val}")
```

```python
## Nested Dictionaries

students = {
    "student1":{"name": "Aditya", "age": 21},
    "student2":{"name": "Adi", "age": 22}
}

print(students)
```

```python
## Accessing nested dictionaries

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## Iterating over nested dictionaries

for student_id, student_info in students.items():
    print(f"Student Id -> {student_id} : Student Info -> {student_info}")
    for key,value in student_info.items():
        print(f"Key -> {key} : Value -> {value}")
```

```python
## Dictionary Comprehension
squares = {x:x**2 for x in range (5)}
print(squares)
```

```python
## conditional Dictionary Comprehension
evens = {x:x**2 for x in range (10) if x%2 == 0}
print(evens)
```

```python
### Practicaal Examples
## Use a dictionary to count the frequency of elements in list

numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
frequency = {}
for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

print(frequency)
```

```python
## Merge 2  dictionaries into one

dict_1 = {"a":1, "b":2}
dict_2 = {"c":3, "d":4}
merged_dict = {** dict_1, ** dict_2}
print(merged_dict)
```
