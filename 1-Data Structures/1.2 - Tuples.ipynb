### Tuples
    Tuples are ordered collection of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
## creating a tupple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
## we can also use methods to create list or tuple
lst = list()
print(type(lst))

tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1, 2, 3, 4, 5, 6])
numbers
```

```python
list((1, 2, 3, 4, 5, 6))
```

```python
mixed_tuple = (1, "Hello World", 3.14, True)
print(mixed_tuple)
```

```python
## accessing tuple elements

numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
numbers[::-2]
```

```python
## Tuples operation

concatenation_tuple = numbers + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple * 3 ## entire tuple appended 3 times
```

```python
numbers * 3
```

```python
## Immutable natures of tuple
## i.e., Tuples are immutable, meaning their element cannot be changed once assigned

lst = [1, 2, 3, 4, 5]
print(lst)

lst[1] = "Aditya"
print(lst)
```

```python
numbers[1]
```

```python
numbers[1] = "Aditya" ## hence proved that tuples are immutable
```

```python
## Tuple Methods

print(numbers.count(1)) ## count the occurrence of mentioned number in the tuple
print(numbers.index(3)) ## returns the first occurrence of index number of the element mentioned
```

```python
## Packing and Unpacking tuple

## packing
packed_tuple = 1, "Hello", 3.14 ## whenever we define separated by comma by default it packs them into tuple
print(packed_tuple)
```

```python
## unpacking
a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *

numbers = (1, 2, 3, 4, 5, 6)
first, *middle, last = numbers
print(first)
print(middle) ## takes all the element from first to last
print(last)
```

```python
## Nested Tuple

lst = [[1, 2, 3, 4], [5, 6, 7, 8], ["Hello", 3.14]] ## example of nested list i.e, list inside list
lst[0][0:3]
```

```python
lst = [[1, 2, 3, 4], [5, 6, 7, 8], ("Hello", 3.14, 'c')] ## mixture list and tuple inside list
lst[2][0:3]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False)) ## example of nested tuple i.e, tuple inside tuple

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## Iterating over nested tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ") ## ends with blank space after each print
    print() ## print new line
```
