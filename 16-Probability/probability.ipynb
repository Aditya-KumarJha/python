# Probability

## Introduction: It's about determining the likelihood of an event.

## Mutual Exclusive Event

### Two events are said to be mutually exclusive if they cannot occur at the same time. For example, if we roll a die, the event of getting a 1 and the event of getting a 2 are mutually exclusive because we cannot get both 1 and 2 at the same time.
### Formula: P(A OR B) = P(A) + P(B) { Additive Rule for Mutual Exclusive Events }

## Non Mutual Exclusive Event

### Two events are said to be non mutually exclusive if they can occur at the same time. For example, if we draw a card from a deck, the event of getting a red card and the event of getting a heart are non mutually exclusive because we can get both a red card and a heart at the same time.
### Formula: P(A OR B) = P(A) + P(B) - P(A AND B) { Additive Rule for Non Mutual Exclusive Events }

## Independent Events
### 2 events are said to be independent if the occurrence of one event does not affect the occurrence of the other event. For example, if we flip a coin twice, the event of getting a head on the first flip and the event of getting a head on the second flip are independent because the outcome of the first flip does not affect the outcome of the second flip.
### Formula: P(A AND B) = P(A) × P(B) { Multiplicative Rule for Independent Events }

## Dependent Events
### 2 events are said to be dependent if the occurrence of one event affects the occurrence of the other event. For example, if we draw a card from a deck and do not replace it, the event of getting a red card on the first draw and the event of getting a red card on the second draw are dependent because the outcome of the first draw affects the outcome of the second draw.
### Formula: P(A AND B) = P(A) × P(B|A), and P(B|A) = P(A AND B) / P(A) { Multiplicative Rule for Dependent Events } it's also known as conditional probability.

## Probability Distribution Function

### Probability Distribution Function (PDF) describes how the probabilities are distributed over the values of a random variable. Two main types of PDF are:
- Probability Mass Function (PMF) for discrete random variables
- Probability Density Function (PDF) for continuous random variables

### Probability Mass Function (PMF): It is a function that gives the probability that a discrete random variable is exactly equal to some value.

### Cumulative Density Function (CDF): It is a function that gives the probability that a random variable is less than or equal to some value.

### Probability Density Function (PDF): It is a function that gives the probability that a continuous random variable is exactly equal to some value.


#### Probability Density = Gradient of Cumulative Density Function (CDF).

### PDF Properites:
- Non negative f(x) >= 0 for all x.
- Total area under the curve is 1.
- P(a <= X <= b) = integral from a to b of f(x)dx
$$\int_{-\infty}^{\infty} f(x)\,dx = 1$$

### Types of Probability Distributions
- Bernoulli Distribution -> Outcomes are 0 or 1. We will use (PMF - Probability Mass Function) for this.
- Binomial Distribution -> It is a generalization of Bernoulli Distribution. We will use (PMF - Probability Mass Function) for this.
- Normal / Gaussian Distribution -> This distribution is used for continuous random variables. We will use (PDF - Probability Density Function) for this.
- Poisson Distribution -> This distribution is used for counting the number of events that occur in a fixed interval of time or space. We will use (PMF - Probability Mass Function) for this.
- Log Normal Distribution -> This distribution is used for continuous random variables. We will use (PDF - Probability Density Function) for this.
- Uniform Distribution -> This distribution is used for continuous random variables. We will use (PDF - Probability Density Function) for this.

#### PDF is for continuous random variables and PMF is for discrete random variables.

#### Bernoulli Distribution

##### The Bernoulli distribution is the simplest discrete probability distribution. It represents the probability distribution of a random variable that has exactly two possible outcomes, usually called "success" (with probability p) and "failure" (with probability 1-p). It is used to model binary outcomes, such as coin flips, yes/no questions, or pass/fail scenarios.
- Entire distribution is for discrete random variable (PMF).
- Outcomes are binary (0 or 1).

###### Parameters
- p: probability of success (0 ≤ p ≤ 1)
- 1-p: probability of failure (0 ≤ 1-p ≤ 1)
- k = {0, 1} => 2 outcomes
$$P(X=k)=p^k(1-p)^{1-k}, \quad k\in\{0,1\}$$

- Mean on Bernoulli = p
$$E[X]=\sum_{k=0}^{1}k\,P(X=k)=0(1-p)+1(p)=p$$
- Variance on Bernoulli = p(1-p)
$$Var(X) = p(1-p)$$
- Median on Bernoulli = 0 if p < 0.5, 1 if p > 0.5, and 0.5 if p = 0.5
$$Median(X) = \begin{cases} 0 & \text{if } p < 0.5 \\ 1 & \text{if } p > 0.5 \\ 0.5 & \text{if } p = 0.5 \end{cases}$$
